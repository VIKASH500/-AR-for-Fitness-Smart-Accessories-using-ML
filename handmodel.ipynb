```python
import gradio as gr
import cv2
import mediapipe as mp

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()

def hand_tracking(frame):
    # Convert the BGR image to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process the frame with MediaPipe Hands
    results = hands.process(rgb_frame)

    # If hands are detected, print a message and return the original frame
    if results.multi_hand_landmarks:
        print("Human hand is detected")

    return frame

# Define the Gradio interface
iface = gr.Interface(
    fn=hand_tracking,
    inputs=gr.Image(type="numpy", label="Back Camera"),
    outputs=gr.Image(type="numpy"),
    live=True,
    title="Live Hand Tracking",
    description="Real-time hand tracking using the back camera.",
)

# Launch the Gradio interface
iface.launch()
```
