```python
import gradio as gr
import pandas as pd
# Load the data from the CSV file
products_df = pd.read_csv('smartwatches.csv')

# List of distinct brand names
distinct_brands = ['noise', 'fire-boltt', 'boat', 'honor', 'crossbeats', 'samsung', 'garmin',
                   'huawei', 'dizo', 'gizmore', 'ambrane', 'zebronics', 'pebble', 'hammer', 'apple',
                   'fitbit', 'amazfit', 'fossil']

# Create a dictionary for Radio component options
brand_options = {brand: brand for brand in distinct_brands}

def search_products(brand, price_range):
    brand = brand.lower()
    price_range = price_range.lower()

    filtered_products = products_df[
        (products_df['Brand'].str.lower().str.contains(brand)) &
        (products_df['Current Price'].astype(str).str.contains(price_range))
    ]

    # Select the top result (you can modify this based on your requirements)
    result = filtered_products.head(1)

    # Extract key features (customize this based on your actual column names)
    key_features = result[['Touchscreen', 'Battery Life (Days)', 'Bluetooth', 'Display Size', 'Weight']].to_dict('records')

    # Return brand, price, rating, battery life, and key features
    return brand.capitalize(), result['Current Price'].values[0], result['Rating'].values[0], result['Battery Life (Days)'].values[0], key_features

# Custom HTML for permanent orange background
custom_html = """
<style>
body {
    background-color: orange;
}
</style>
"""

iface = gr.Interface(
    fn=search_products,
    inputs=[gr.Radio(brand_options), "text"],
    outputs=["text", "text", "text", "text", "text"],
    live=True,
    title="Smartwatches Search Interface",
    description="Choose brand and enter price range to search for smartwatches.",
)

iface.launch()
```
